```python
#Histogram of text length for entire text column
text_length = symp.text.apply(lambda x: len(x.split(' ')))

sns.histplot(text_length, color= 'teal', kde=True, bins= 20)
plt.xlabel('Length of Text')
plt.show()
```

```python
# wordcloud of symptom texts

stop_words = set(stopwords.words('english'))
# add some stop words
stop_words.update(["I've", "I'm", "like"])

wordcloud = WordCloud(width = 800, 
                      height = 800, 
                      background_color ='white', 
                      stopwords=stop_words, 
                      min_font_size = 8).generate(str(symp.text))

# display the word cloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
