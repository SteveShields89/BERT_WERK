### Preprocessing workflow:
    1. Label encoder for diseases
    2. Use of pycaret preprocessing to remove stop words and embedding using (Latent Dirichlet Allocation)
    3. Tokenize words so they can be put in to models

```python
#Label Encoding
def label_encode(labels):
    label_encoder = LabelEncoder()
    label_encoder.fit(labels)
    label_sequences = label_encoder.transform(labels)
    return label_sequences


symp['encoded'] = label_encode(symp['label'])
```

```python
def preprocess_text_with_padding(df, text_column, processed_column, max_length):
    # Initialize lemmatizer
    lemmatizer = WordNetLemmatizer()

    # Remove punctuation and convert to lowercase
    df[processed_column] = df[text_column].apply(lambda x: re.sub(r'[^\w\s]', '', x.lower()))

    # Tokenize text
    df[processed_column] = df[processed_column].apply(lambda x: nltk.word_tokenize(x))

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    df[processed_column] = df[processed_column].apply(lambda x: [word for word in x if word not in stop_words])

    # Lemmatize words
    df[processed_column] = df[processed_column].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])

    # Convert words to sequences of integers
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(df[processed_column])
    sequences = tokenizer.texts_to_sequences(df[processed_column])

    # Pad sequences
    padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')

    # Store the padded sequences in the processed column
    df[processed_column] = padded_sequences.tolist()

    return df
```

```python
# Preprocess text and create a new column called 'Processed' with padding
max_length = 50
model_df = preprocess_text_with_padding(symp, 'text', 'processed', max_length)
```

```python
X =model_df.processed
y= model_df.encoded
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, stratify = y)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Modeling will consists of only 3 models:
    
    1. Multinomial NB
    2. Logistic Regression
    3. Random Forest

```python
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
log =LogisticRegression()
log.fit(X_train, y_train)
print('train score:', log.score(X_train, y_train))
print('test score:', log.score(X_test, y_test))
```

```python
tfidif = TfidfVectorizer(analyzer= 'word', stop_words='english',max_features=10000)
nb = MultinomialNB()

pipe_nb = Pipeline([
    ('tfidif', tfidif),
    ('nb',nb)
])
pipe_nb.fit(X_train, y_train)
print('train score:', pipe_nb.score(X_train, y_train))
print('test score:', pipe_nb.score(X_test, y_test))
```

```python
preds_nb= pipe_nb.predict(X_test)
test2 = pd.concat([X_test,y_test], axis =1)
test2['predicted'] = preds_nb
test2
```

#### Using CountVectorizer with logistic regression using a pipeline:

```python
cvec = CountVectorizer(analyzer = "word",                        
                        stop_words = "english", 
                         max_features = 10000)

lr = LogisticRegression()

pipe_lr = Pipeline([
    ('cvec', cvec),
    ('lr', lr)
])

pipe_lr.fit(X_train, y_train)
print('train score:', pipe_lr.score(X_train, y_train))
print('test score:', pipe_lr.score(X_test, y_test))
```

```python
# Get predictions
preds_lr= pipe_lr.predict(X_test)
preds_lr
test = pd.concat([X_test,y_test], axis =1)
test['predicted'] = preds_lr
test
```

```python
rf =RandomForestClassifier()
tfidif = TfidfVectorizer(analyzer='word', stop_words='english', max_features=10000)
pipe_rf = Pipeline([
    ('tfidif', tfidif),
    ('rf',rf)])

pipe_rf.fit(X_train, y_train)
print('train score:', pipe_rf.score(X_train, y_train))
print('test score:', pipe_rf.score(X_test, y_test))
```

```python
preds_rf= pipe_rf.predict(X_test)
preds_rf
test3 = pd.concat([X_test,y_test], axis =1)
test3['predicted'] = preds_rf
test3
```
